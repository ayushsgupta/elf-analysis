```python
def round_sig(x, sig=3):
    return round(x, sig-int(floor(log10(abs(x))))-1)
```

```python
class EnhancedVD(VolumetricData):
    
    def __init__(self, structure, data, distance_matrix=None, data_aug=None):
        super().__init__(structure, data, distance_matrix, data_aug)

        self.xpoints = np.linspace(0.0, 1.0, num=self.dim[0])
        self.ypoints = np.linspace(0.0, 1.0, num=self.dim[1])
        self.zpoints = np.linspace(0.0, 1.0, num=self.dim[2])
        
        self.interpolator = RegularGridInterpolator((self.xpoints, self.ypoints, self.zpoints), self.data['total'], bounds_error=True)
    
    def val_at(self, x, y, z):
        #assert min(x, y, z) >= 0.0 and max(x, y, z) <= 1.0 #Don't need if bounds_error=True on self.interpolator
        result, interp = self._interpolate(x, y, z)
        if interp:
            # Raise some warning?
            pass
        return result
        
    def _interpolate(self, x, y, z):
        interpolation_used = not (x in self.xpoints and y in self.ypoints and z in self.zpoints)
        return self.interpolator([x, y, z])[0], interpolation_used
    
    def linear_slice(self, p1, p2, n=100):
        assert type(p1) in [list, np.ndarray] and type(p2) in [list, np.ndarray]
        assert len(p1) == 3 and len(p2) == 3
        xpts = np.linspace(p1[0], p2[0], num=n)
        ypts = np.linspace(p1[1], p2[1], num=n)
        zpts = np.linspace(p1[2], p2[2], num=n)
        return [self._interpolate(xpts[i], ypts[i], zpts[i])[0] for i in range(n)]
    
    def planar_slice(self, miller_indices, xspacing=0.1, yspacing=0.1):
        h, k, l = miller_indices
        
    def get_integrated_total(self, ind, radius, nbins=1):
        # For non-spin-polarized runs, this is zero by definition.
        if not self.is_spin_polarized:
            radii = [radius / nbins * (i + 1) for i in range(nbins)]
            data = np.zeros((nbins, 2))
            data[:, 0] = radii
            return data

        struct = self.structure
        a = self.dim
        if ind not in self._distance_matrix or \
                self._distance_matrix[ind]["max_radius"] < radius:
            coords = []
            for (x, y, z) in itertools.product(*[list(range(i)) for i in a]):
                coords.append([x / a[0], y / a[1], z / a[2]])
            sites_dist = struct.lattice.get_points_in_sphere(
                coords, struct[ind].coords, radius)
            self._distance_matrix[ind] = {"max_radius": radius,
                                          "data": np.array(sites_dist)}

        data = self._distance_matrix[ind]["data"]

        # Use boolean indexing to find all charges within the desired distance.
        inds = data[:, 1] <= radius
        dists = data[inds, 1]
        data_inds = np.rint(np.mod(list(data[inds, 0]), 1) *
                            np.tile(a, (len(dists), 1))).astype(int)
        vals = [self.data["total"][x, y, z] for x, y, z in data_inds]

        hist, edges = np.histogram(dists, bins=nbins,
                                   range=[0, radius],
                                   weights=vals)
        data = np.zeros((nbins, 2))
        data[:, 0] = edges[1:]
        data[:, 1] = [sum(hist[0:i + 1]) / self.ngridpts
                      for i in range(nbins)]
        return data
    
    def get_avg_data_val(self, index, r=0.4):
        isite = self.structure.sites[index]
        mask = self._dist_mat(isite.frac_coords) < r
        vol_sphere = self.structure.volume * (mask.sum() / self.ngridpts)
        avg_data_val = np.sum(self.data['total'] * mask) / mask.size / vol_sphere
        return avg_data_val

    def _dist_mat(self, pos_frac):
        # return a matrix that contains the distances
        aa = np.linspace(0, 1, len(self.get_axis_grid(0)),
                         endpoint=False)
        bb = np.linspace(0, 1, len(self.get_axis_grid(1)),
                         endpoint=False)
        cc = np.linspace(0, 1, len(self.get_axis_grid(2)),
                         endpoint=False)
        AA, BB, CC = np.meshgrid(aa, bb, cc, indexing='ij')
        #print('POS FRAC', pos_frac)
        dist_from_pos = self.structure.lattice.get_all_distances(
            fcoords1=np.vstack([AA.flatten(), BB.flatten(), CC.flatten()]).T,
            fcoords2=pos_frac)
        return dist_from_pos.reshape(AA.shape)
        
class NewElfcar(EnhancedVD):
    
    def __init__(self, poscar, data):
        super().__init__(poscar.structure, data)
        #self.data = data

    @classmethod
    def from_file(cls, filename):
        (poscar, data, data_aug) = VolumetricData.parse_file(filename)
        return cls(poscar, data)
    
    @classmethod
    def from_old_elfcar(cls, elfcar):                
        return cls(elfcar, elfcar.data)

    def get_alpha(self):
        alpha_data = {}
        for k, v in self.data.items():
            alpha = 1 / v
            alpha = alpha - 1
            alpha = np.sqrt(alpha)
            alpha_data[k] = alpha
        return EnhancedVD(self.structure, alpha_data)
```

```python
def get_bonds(structure, tol=0.5):
    bonds = []
    for site1, site2 in itertools.combinations(structure.sites, 2):
        if CovalentBond.is_bonded(site1, site2, tol, default_bl=2.1):
            bonds.append(CovalentBond(site1, site2))
    return bonds
```

```python
def get_bonds_nn(structure):
    bonds = []
    for i in range(len(structure.sites)):
        other_sites = bond_nn.get_nn_info(structure, i)
        for other_site in other_sites:
            bond = CovalentBond(structure.sites[i], other_site['site'])
            bonds.append(bond)
    return bonds
```

```python
def get_geos_nn(structure):
    bonds = []
    for i in range(len(structure.sites)):
        other_sites = geo_nn.get_nn_info(structure, i)
        for other_site in other_sites:
            bond = CovalentBond(structure.sites[i], other_site['site'])
            bonds.append(bond)
    return bonds
```

```python
def plot_all_bonds(f, bond_list, plot=False, mode='elf', savefig=False, return_avg=False):
    traversed, master = [], []
    cation, anion = Composition(f).elements
    disp = 'ELF' if mode == 'elf' else 'α'
    num_bonds = 0
    if not bond_list:
        #print('BAD BOY!', f)
        return None
    for bond in bond_list:
        try: 
            coords = [list(bond.site1.frac_coords), list(bond.site2.frac_coords)]
            cond = any([t == coords or t == coords[::-1] for t in traversed])
            if cond:
                #print('DUPLICATE!')
                continue
            if str(bond.site1.specie) != str(cation):
                assert str(bond.site1.specie) == str(anion)
                coords.reverse()
            for i in range(3):
                coords[0][i] = 0.0 if abs(coords[0][i]) < 1e-3 else coords[0][i]
                coords[1][i] = 0.0 if abs(coords[1][i]) < 1e-3 else coords[1][i]
            if mode == 'elf':
                data = elfcar_dict[f].linear_slice(coords[0], coords[1])
            elif mode == 'alpha':
                data = elfcar_dict[f].get_alpha().linear_slice(coords[0], coords[1], n=1000)
            master.append(data)
            traversed.append(coords)
            num_bonds += 1
            #print('YIPPEE!', f, coords)
        except ValueError:
            pass
            #print('BOO HOO!', f, coords)
    
    if plot:
        for data in master:
            plt.plot(data)
        plt.title('{0} {1}, {2} eV error'.format(f, disp, str(round_sig(error_dict[f]))), fontsize=18)
        plt.tick_params(axis='x', which='both', bottom=False, top=False, labelbottom=False)
        plt.xlabel('Cation -------------------({0} bonds)---------------------> Anion'.format(num_bonds), fontsize=16)
        plt.ylabel(disp, fontsize=16)
        if mode == 'alpha':
            plt.yscale('log')
        if savefig:
            plt.savefig('linear_slices/{0}/{1}'.format(mode, f), dpi=300)
        plt.show()
        
    if return_avg:
        nps = np.array(master)
        return np.mean(nps, axis=0)
```

```python
def the_whole_shebang(f, plot=False, mode='elf', savefig=False, which='bond', return_avg=False):
    struct = elfcar_dict[f].structure
    if which=='bond':
        bonds = get_bonds_nn(struct)
    else:
        bonds = get_geos_nn(struct)
    #coords = extract_coordinates(bonds)
    return plot_all_bonds(f, bonds, plot, mode, savefig, return_avg)
```

```python
def rdf(f, plot=True, mode='elf', savefig=False, di=False, spinup=True):
    data_list = []
    cation, anion = Composition(f).elements
    disp = 'RDF (ELF)' if mode == 'elf' else 'RDF (α)'
    if mode == 'elf':
        elfcar = elfcar_dict[f]
    elif mode == 'alpha':
        elfcar = elfcar_dict[f].get_alpha()
    struct = elfcar.structure
    for i in range(len(struct.sites)):
        #print(struct.sites[i])
        if spinup:
            dat = elfcar.get_integrated_total(i, 2, nbins=50)
        else:
            dat = elfcar.get_integrated_diff(i, 2, nbins=50)
        #print(dat)
        dat = [[d[0], d[1]] for d in dat]
#         dat2 = [list(dat[0]), list(dat[1])]
        data_list.append(dat)
    
    def deintegrate(lst):
        newlst = [lst[0][1]]
        for i in range(1, len(lst)):
            newlst.append(lst[i][1] - lst[i - 1][1])
        return newlst
    
    if di:
        new_data = [deintegrate(d) for d in data_list]
    else:
        #new_data = [1, 2, 3, 4]
        new_data = [[d[1] for d in d1] for d1 in data_list]
    #print(new_data)
    
    if plot:
        for data in new_data:
            plt.plot(data)
        plt.title('{0} {1}, {2} eV error'.format(f, disp, str(round_sig(error_dict[f]))), fontsize=18)
        plt.tick_params(axis='x', which='both', bottom=False, top=False, labelbottom=False)
        #plt.xlabel('Cation -------------------({0} bonds)---------------------> Anion'.format(num_bonds), fontsize=16)
        plt.ylabel(disp, fontsize=16)
#         if mode == 'alpha':
        if False:
            plt.yscale('log')
        if savefig:
            plt.savefig('{1}_{0}_rdf'.format(mode, f), dpi=300)
        plt.show()
        
    #return data_list
```

```python
def spherical_avg(f, mode='elf', savefig=False, sp=False):
    data_list = []
    cation, anion = Composition(f).elements
    disp = 'RDF (ELF)' if mode == 'elf' else 'RDF (α)'
    if mode == 'elf':
        elfcar = elfcar_dict[f]
    elif mode == 'alpha':
        elfcar = elfcar_dict[f].get_alpha()
    struct = elfcar.structure
    latt = struct.lattice
    max_r = min(latt.a, latt.b, latt.c) / 2
    pts = np.linspace(0.0, 2, num=1000)
    for i in range(len(struct.sites)):
        dat = []
        for r1 in pts:
            dat.append(elfcar.get_avg_data_val(i, r1))
        data_list.append(dat)

    for data in data_list:
        plt.plot(pts, data)
    if mode == 'alpha':
        plt.yscale('log')
    plt.title('{0} {1}, {2} eV error'.format(f, disp, str(round_sig(error_dict[f]))), fontsize=18)
    plt.xlabel('Distance from atom (Å)', fontsize=16)
    plt.ylabel(disp, fontsize=16)
    
    if savefig: 
        if sp:
            plt.savefig('special_{1}_{0}_rdf'.format(mode, f), dpi=300)
        else:
            plt.savefig('spherical_rdf/{1}_{0}_rdf'.format(mode, f), dpi=300)
        print(f, 'saved!')
        
    plt.show()
```

```python
cl_spg, cl_lat, cl_form, cl_conf, cl_confa = [], [], [], [], []

roundtup = lambda a, b, c: tuple([round(num, 3) for num in [a, b, c]])

for f in chlorides + chlorides3:
    cation, anion = Composition(f).elements
    struct = elfcar_dict[f].structure
    cl_spg.append(struct.get_space_group_info())
    cl_lat.append(roundtup(struct.lattice.a, struct.lattice.b, struct.lattice.c))
    cl_form.append(expt[f])
    cl_conf.append(re.sub("</*sup>", "", cation.electronic_structure))
    cl_confa.append(re.sub("</*sup>", "", anion.electronic_structure))
    
chloride_data = pd.DataFrame.from_dict({'Formula':chlorides+chlorides3, 'Formation_Energy':cl_form, 'Spacegroups':cl_spg, 'Lattice':cl_lat, 'Electronic_Structure_Cation':cl_conf, 'Electronic_Structure_Anion':cl_confa})
```
