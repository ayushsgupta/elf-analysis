```python
def reorganize_files(): 
    print('Reorganizing for:')
    count, empty = 0, 0
    # loop through the directories containing the calcs
    block_dir = data_dir
    
    for d1 in block_dir.iterdir():
        if '.DS_Store' in str(d1):
            empty += 1
            continue
        
        # loop through the "launcher" dirs in each "block"
        for d2 in d1.iterdir():
            count+=1
            
            if '.DS_Store' in str(d2):
                empty += 1
                continue
                
            def file_helper(poscar_path, chgcar_path, aeccar0_path, aeccar2_path):
                formula = Structure.from_file(poscar_path).composition.reduced_formula
                compounds_all.append(formula)
                print('---------------------------------------------------------------')
                print(formula)
                
                if not os.path.isdir('vasp_files/by_compound/{}/'.format(formula)):
                    subprocess.run(['mkdir', 'vasp_files/by_compound/{}/'.format(formula)])
                    print('\tFolder for', formula, 'successfully made!')
                else:
                    print('\tFolder for', formula, 'already exists!')
                    
                target_dir = output_dir / 'by_compound' / formula
                
                if not os.path.isfile('vasp_files/by_compound/{}/CHGCAR.vasp'.format(formula)):           
                    chgcar = Chgcar.from_file(chgcar_path)
                    chgcar.write_file(Path(target_dir, 'CHGCAR.vasp'))
                    print('\tCHGCAR written') 
                else:
                    print('\tCHGCAR already exists')
                if not os.path.isfile('vasp_files/by_compound/{}/CHGREF.vasp'.format(formula)):           
                    aeccar0 = Chgcar.from_file(aeccar0_path)
                    aeccar2 = Chgcar.from_file(aeccar2_path)
                    chgref = aeccar0 + aeccar2
                    chgref.write_file(target_dir / 'CHGREF.vasp')
                    print('\tCHGREF written')
                else:
                    print('\tCHGREF already exists')
                
            try:
                poscar_path = d2 / 'POSCAR.relax2.gz'
                chgcar_path = d2 / 'CHGCAR.relax2.gz'
                aeccar0_path = d2 / 'AECCAR0.relax2.gz'
                aeccar2_path = d2 / 'AECCAR2.relax2.gz'    
                file_helper(poscar_path, chgcar_path, aeccar0_path, aeccar2_path)
            except FileNotFoundError as e:
                poscar_path = d2 / 'POSCAR.relax3.gz'
                chgcar_path = d2 / 'CHGCAR.relax3.gz'
                aeccar0_path = d2 / 'AECCAR0.relax3.gz'
                aeccar2_path = d2 / 'AECCAR2.relax3.gz'
                file_helper(poscar_path, chgcar_path, aeccar0_path, aeccar2_path)
#             except TypeError as e:
#                 print('AECCAR, you naughty boy!')
#                 print(e)
#             except Exception as exc:
#                 print({exc})
#                 continue
        
    print('---------------------------------------------------------------')    
    print('Done! Traversed {0} directories, of which {1} were empty'.format(count, empty))
```

```python
compounds = ['TiC', 'IrO2', 'TaN', 'TiO2', 'MnBr2', 'Mn3O4', 'Pt5Se4', 'Fe2O3']
```

```python
def run_bader():
    target_dir = output_dir / 'by_compound'
    for f in compounds:
        working_dir = target_dir / f
        os.chdir(working_dir)
        if not os.path.isfile('Bvol0001.dat'):
            subprocess.run(['bader', '-p', 'atom_index', 'CHGCAR.vasp', '-ref', 'CHGREF.vasp'])
            print('Bader complete', f)
        else:
            print('Bader already ran', f)
```

```python
run_bader()
```

```python
def retrieve_bader_results(return_dict={}):
    target_dir = output_dir / 'by_compound'
    for f in compounds:
        if not return_dict.get(f, None):
            chgcars = []
            working_dir = target_dir / f
            count = 0
            for entry in working_dir.iterdir():
                if re.search(r'Bvol', str(entry)):
                    chgcars.append(Chgcar.from_file(entry))
                    count += 1
            return_dict[f] = chgcars
            print(f, 'done:', count, 'files')
        else:
            print(f, 'already done')
```

```python
chgcars_dict = {}
```

```python
retrieve_bader_results(chgcars_dict)
```

```python
def filter_zeros(chgcars):
    filtered = {}
    for f in chgcars.keys():
        lst, lst2 = chgcars[f], []
        for chg in lst:
            data = chg.data['total'].flatten()
            data = data[data != 0.0]
            lst2.append(data)
        filtered[f] = lst2
    return filtered
```

```python
filtered_chgcars = filter_zeros(chgcars_dict)
```

```python
def describe_stats(f):
    dat = filtered_chgcars[f]
    nums, means, var, ranges = [], [], [], []
    print(f)
    for thing in dat:
        res = describe(thing)
        nums.append(res.nobs)
        means.append(res.mean)
        var.append(res.variance)
        ranges.append(res.minmax[1] - res.minmax[0])
    print('Number of datapoints', nums)
    print('Means', means)
    print('Variances', var)
    print('Ranges', ranges)
    print('-----------------------------------------------------------')
```

```python
for f in compounds:
    describe_stats(f)
```

```python
def create_mask(partial_chgcar):
    dt = partial_chgcar.data['total']
    shape = dt.shape
    flat = dt.flatten()
    for i in range(len(flat)):
        flat[i] = 1.0 if flat[i] != 0.0 else 0.0
    mask = flat.reshape(shape)
    return mask
```

```python
for fm in compounds:
    print(fm)
    sm = 0.0
    for chg in chgcars_dict[fm]:
        msk = create_mask(chg)
        sm += np.average(msk.flatten())
    print(sm)
```

```python
def apply_mask(elfcar, mask):
    dt = elfcar.data['total']
    s1, s2 = dt.shape, mask.shape
    factor = s2[0] / s1[0]
    flat, mf = dt.flatten(), mask.flatten()
    for i in range(len(flat)):
        flat[i] = flat[i] * mf[int(i*factor)]
    masked = flat.reshape(s1)
    return masked
```

```python
iro2 = chgcars_dict['IrO2'][0]
elfc = elfcar_dict['IrO2']
```

```python
mask = create_mask(iro2)
masked_elf = apply_mask(elfc, mask)
```

```python
def create_mask_dict():
    elf, alpha = {}, {}
    for f in compounds:
        try:
            el, al = elfcar_dict[f], elfcar_dict[f].get_alpha()
            masked_e, masked_a = [], []
            for chg in chgcars_dict[f]:
                mask = create_mask(chg)
                masked_e.append(apply_mask(el, mask))
                masked_a.append(apply_mask(al, mask))
            elf[f], alpha[f] = masked_e, masked_a
        except KeyError:
            print(f, 'has no ELFCAR')
    return elf, alpha
```

```python
masked_elf, masked_alpha = create_mask_dict()
```

```python
masked_elf.keys(), masked_alpha.keys()
```

```python
def plot_histogram(f, mode='elf'):
    if mode == 'elf':
        data = masked_elf[f]
        disp = 'ELF'
    elif mode == 'alpha':
        data = masked_alpha[f]
        disp = ''
        
    processed = []
    for dat in data:
        processed.append([dp for dp in dat.flatten() if dp != 0.0])
    
        
    nbins = 100
    plt.hist(processed, 
             nbins, 
             histtype='bar', 
             stacked=True,
             label=['Bader Well {}'.format(i + 1) for i in range(len(dat))],
    )
    #plt.legend(loc="upper right")
    if mode == 'alpha':
        plt.yscale('log')
    
    plt.title('Distribution of {0} Values Across Bader Wells for {1}'.format(disp, f), fontsize=20)
    plt.xlabel('{} Value'.format(disp), fontsize=16)
    plt.ylabel('Count', fontsize=16)

    plt.show()
```

```python
for f in masked_elf.keys():
    plot_histogram(f, mode='elf')
    plot_histogram(f, mode='alpha')
```
